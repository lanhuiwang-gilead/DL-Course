# Examples of Image Convolution

Lets look at what happens when the following matrixes are 'convolved' or applied as 2D filters to the matrix that represents the cat picture.

```python
# Try this yourself at http://setosa.io/ev/image-kernels/
```

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
# The box blur filter(https://en.wikipedia.org/wiki/Box_blur#:~:text=A%20box%20blur%20(also%20known,(%22blurring%22)%20filter.)
blur = np.array((
	[1, 1, 1],
	[1, 1, 1],
	[1, 1, 1]), dtype="int") * 1.0/9.0

# construct the Sobel x-axis kernel
sobelX = np.array((
	[-1, 0, 1],
	[-2, 0, 2],
	[-1, 0, 1]), dtype="int")
 
# construct the Sobel y-axis kernel
sobelY = np.array((
	[-1, -2, -1],
	[0, 0, 0],
	[1, 2, 1]), dtype="int")

laplacian = np.array((
	[0, 1, 0],
	[1, -4, 1],
	[0, 1, 0]), dtype="int")
```

```python
# source image

img = cv2.imread('cat.jpeg',0)

plt.imshow(img, cmap = 'gray', interpolation = 'bicubic')
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show()
```

```python
# blur using box filter blur
kernel = blur

dst = cv2.filter2D(img,-1,kernel)

plt.imshow(dst, cmap = 'gray', interpolation = 'bicubic')
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show()
```

```python
# sobel filter Y - finds vertical edges
dst = cv2.filter2D(img,-1,sobelY)

plt.imshow(dst, cmap = 'plasma', interpolation = 'bicubic')
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show()
```

```python
# sobel filter X - finds horizontal edges
dst = cv2.filter2D(img,-1,sobelX)

plt.imshow(dst, cmap = 'plasma', interpolation = 'bicubic')
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show()
```

```python
# laplacian - finds areas of most change but very sensitive to noise
dst = cv2.filter2D(img,-1,laplacian)

plt.imshow(dst, cmap = 'plasma', interpolation = 'bicubic')
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show()
```

```python
# laplacian of guassian blurred image
gaussian1d = cv2.getGaussianKernel(18,-1)
dst = cv2.sepFilter2D(img,-1,gaussian1d,gaussian1d)
lp = cv2.filter2D(dst,-1,laplacian)

plt.imshow(lp, cmap = 'plasma', interpolation = 'bicubic')
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show()
```

```python
gaussian1d = cv2.getGaussianKernel(18,-1)
dst = cv2.sepFilter2D(img,-1,gaussian1d,gaussian1d)
lp = cv2.filter2D(dst,-1,laplacian)
sharp = img - lp
plt.imshow(sharp, cmap = 'gray', interpolation = 'bicubic')
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show()
```

```python
gaussian1d = cv2.getGaussianKernel(18,-1)
dst = cv2.sepFilter2D(img,-1,gaussian1d,gaussian1d)
plt.imshow(dst, cmap = 'gray', interpolation = 'bicubic')
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show()
```
